LINEAR REGRESSION WITH GRID SEARCH CV

```python
# Linear Regression
import pandas as pd 
import seaborn as sns 
df=sns.load_dataset("titanic")
df['age']=df['age'].fillna(value=df['age'].mean())
X=df[['age']]
y=df['fare']

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

# Create a Model / Hyperparameter Tuning

model= LinearRegression()

# Define parameter grid
param_grid = {'fit_intercept':[True,False]}

#object grid search cv (Creating the model)
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='r2')

#training the model
grid_search.fit(X,y)

#print the best parameters
print("Best Parameters: ", grid_search.best_params_)
```

# KNN GRID SEARCH CV

```python
# KNN
import pandas as pd 
import seaborn as sns 
df=sns.load_dataset("titanic")

X=df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y=df['survived']
X=pd.get_dummies(X, columns=['sex'])

X.age.fillna(value=X['age'].mean(),inplace=True)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Create a Model / Hyperparameter Tuning

model= KNeighborsClassifier()

# Define parameter grid
param_grid = {'n_neighbors': np.arange(1,40,2), 'weights':['uniform','distance']}

#Not use if use numpy
#param_grid = {'n_neighbors':[1,3,5,7,8,9], 'weights':['uniform','distance']}


#object grid search cv (Creating the model)
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='precision')

#training the model
grid_search.fit(X,y)

#print the best parameters
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

# DECISION TREE CLASSIFIER

```python
# DECISION TREE CLASSIFIER
import pandas as pd 
import seaborn as sns
df=sns.load_dataset("titanic")

X=df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y=df['survived']
X=pd.get_dummies(X, columns=['sex'])
X.age.fillna(value=X['age'].mean(),inplace=True)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Create a Model / Hyperparameter Tuning
model= DecisionTreeClassifier()

# Define parameter grid
param_grid = {'max_depth': [3, 5, 7,None], 'min_samples_split':[2, 3, 4]}

#Not use if use numpy
#param_grid = {'n_neighbors':[1,3,5,7,8,9], 'weights':['uniform','distance']}


#object grid search cv (Creating the model)
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='precision')

#training the model
grid_search.fit(X,y)

#print the best parameters
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

# RANDOM FOREST CLASSIFIER

```python
# RANDOM FOREST CLASSIFIER
import pandas as pd 
import seaborn as sns
df=sns.load_dataset("titanic")

X=df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y=df['survived']
X=pd.get_dummies(X, columns=['sex'])
X.age.fillna(value=X['age'].mean(),inplace=True)

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Create a Model / Hyperparameter Tuning
model= RandomForestClassifier()

# Define parameter grid
param_grid = {'n_estimators':[10,50,100],
              'max_depth':[3,5,7,None],
              'min_samples_split':[2,3,4],
              'max_features':['auto', 'sqrt', 'log2']}

#Not use if use numpy
#param_grid = {'n_neighbors':[1,3,5,7,8,9], 'weights':['uniform','distance']}


#object grid search cv (Creating the model)
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='precision')

#training the model
grid_search.fit(X,y)

#print the best parameters
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```
